# What is RT-DETR?

RT-DETR (Real-Time Detection Transformer) is a transformer-based object detection model designed to achieve high detection accuracy while maintaining real-time inference performance.

Traditional DETR models often suffer from slow convergence and high latency. RT-DETR addresses these limitations through efficient encoder-decoder design and optimized feature processing.

Key advantages:

- End-to-end object detection
- No Non-Maximum Suppression (NMS) required
- High accuracy
- Real-time inference capability
- Strong performance on custom datasets

# Loading a Pretrained RT-DETR Model

Using pretrained weights significantly improves training efficiency and model performance.

```python
from ultralytics import RTDETR

model = RTDETR("rtdetr-l.pt")

print("Model loaded successfully")
```

# Training RT-DETR

Training is performed using a custom dataset configuration file.

```python
model.train(
    data="dataset.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device=[0,1,2],
    name="epochs50_rtdetr"
)
```

# Evaluation Metrics

The following metrics are commonly used to evaluate object detection performance:

### mAP (Mean Average Precision)

Measures overall detection accuracy.

### Precision

Measures the proportion of predicted detections that are correct.

### Recall

Measures the proportion of actual objects successfully detected.

A good object detection model should achieve high values for all three metrics.

# Running Inference

After training, the best model weights can be used to make predictions on new images.

```python
model = RTDETR("best.pt")

results = model.predict(
    source="test.jpg",
    save=True,
    conf=0.25
)
```

# Practical Tips

- Ensure annotation quality is consistent.
- Remove blurry and duplicate images.
- Use pretrained weights whenever possible.
- Monitor mAP rather than relying only on training loss.
- Validate models on unseen data before deployment.

In real-world projects, improving dataset quality often yields larger gains than increasing the number of training epochs.
